## Data Visualization

```python
plt.figure(figsize=(8, 6))
sns.histplot(data["Ladder score"], bins=20, kde=True)
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.title('Distribution of Happiness Scores (2021)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=data["Ladder score"], color='red')
plt.xlabel('Happiness Score')
plt.title('Box Plot of Happiness Scores (2021)')
plt.show()
```

From the above two plots, we can conclude that majority of the countries have happiness score between 5 and 6.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Social support', y='Ladder score', data=data)
plt.xlabel('Social support')
plt.ylabel('Happiness Score')
plt.title('Relationship Between Happiness and Social support (2021)')
plt.show()
```

From the above graph we can infer that higher is the social support, higher is the happiness score. It is visualized that most of the countires have the values of social support ranging from 0.8 to 0.9.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Healthy life expectancy', y='Ladder score', data=data)
plt.xlabel('Healthy life expectancy')
plt.ylabel('Happiness Score')
plt.title('Relationship Between Happiness and Healthy life expectancy(2021)')
plt.show()
```

It is observed from the above graph that Healthy life expectancy varies directly with the happiness score.

### Analyzing World Happiness Data and Finding Reasons Behind Happiness

```python
sns.pairplot(data[['Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
                'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']])
plt.suptitle('Pairplot of Happiness Factors (2021)')
plt.show()

# Correlation matrix to quantify relationships
correlation_matrix = data[['Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
                        'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Relationship Between Happiness Factors (Correlation Matrix)')
plt.show()

```

### Analyzing How Different Factors Affect Happiness Score

```python
from sklearn.linear_model import LinearRegression

# Select relevant features and target variable
X = data[['Logged GDP per capita', 'Social support', 'Healthy life expectancy',
        'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
y = data['Ladder score']

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Get coefficients
coefficients = pd.Series(model.coef_, index=X.columns)
print("Regression Coefficients:")
print(coefficients)
```

### Happiest and Unhappiest Countries

```python
happiness_years = ['2021', '2022', '2023']

for year in happiness_years:
    # Filter the data for the specified year
    year_data = df[df['year'] == year]
    
    # Check if there are data points for the specified year
    if not year_data.empty:
        # Get the top 10 and bottom 10 countries by happiness score
        happiest = year_data.nlargest(10, 'Ladder score')
        unhappiest = year_data.nsmallest(10, 'Ladder score')
        
        # Create bar plots for the happiest and unhappiest countries
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Ladder score', y='Country name', data=happiest, palette='coolwarm')
        plt.xlabel('Happiness Score')
        plt.title(f'Happiest Countries in {year}')
        plt.show()
    else:
        print(f"No data available for the year {year}")
```

#### Number of countries in various regions

```python
import plotly.express as px

# Create a DataFrame with counts of each regional indicator
regional_counts = data["Regional indicator"].value_counts().reset_index()
regional_counts.columns = ["Regional indicator", "Count"]

# Create the pie chart
fig = px.pie(regional_counts, values="Count", names="Regional indicator", color_discrete_sequence=px.colors.sequential.Inferno)
fig.update_traces(pull=[0.1, 0, 0, 0], textposition="inside", textinfo="percent+label")
fig.show()
```

###  Ladder Score Distribution by Countries in Map View 

```python
import plotly.express as px
fig = px.choropleth(data, locations='Country name', locationmode='country names', color='Ladder score',
                    hover_name='Country name', color_continuous_scale='Viridis',
                    title='Ladder Score Distribution by Countries (2021)')
fig.show()
```

The choropleth map visually displays the distribution of ladder scores across countries in 2021. It highlights variations in happiness levels globally, with some countries scoring higher, indicating greater well-being, and others scoring lower, suggesting lower happiness levels. This map offers a quick overview of worldwide happiness disparities, providing valuable insights for policymakers and researchers looking to address well-being disparities and understand the factors influencing them.
